```python
import pandas as pd
import os
import random
import matplotlib.pyplot as plt
from src.formation_gui.field import Field
```

```python
# Load datasets
data_files_basepath = f"{os.getcwd()}\\resources"

plays = pd.read_csv(f"{data_files_basepath}\\reduced_data.csv")
```

```python
rand_game = random.choice(plays.gameId.unique())
rand_play = random.choice(plays[plays['gameId'] == rand_game].playId.unique())

play = plays[(plays['gameId'] == rand_game) & (plays['playId'] == rand_play)]
play
```

```python
from src.formation_gui.field import Field

field_viz = Field()

field_viz.generate_field()
field_viz.add_players(play)
```

```python
suffix = ""
if play.playDirection.unique()[0] == 'left':
    fig_l = field_viz.zoomed_formation(play, title_suffix=" - Original")
    play = field_viz.flip_formation(play)
    suffix = " - Flipped"

fig_r = field_viz.zoomed_formation(play, suffix)
```
